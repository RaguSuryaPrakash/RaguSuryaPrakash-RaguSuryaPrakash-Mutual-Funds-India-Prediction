# Does SCHEME vary YEARLY RETURNS?

# We will use violine plot for visualising YEARLY RETURNS 1 TO 5

```python
fig, axs = plt.subplots(1, 3, gridspec_kw={'width_ratios': [5, 3, 3]}, figsize=(25, 5))
sns.violinplot(y="returns_1yr", x="category", data=dataset.sort_values("alpha", ascending=False), ax=axs[0])
axs[0].set_title("Returns (1 Year) by Category", fontsize=20)
sns.violinplot(y="returns_3yr", x="category", data=dataset.sort_values("alpha", ascending=False), ax=axs[1])
axs[1].set_title("Returns (3 Years) by Category", fontsize=20)
sns.violinplot(y="returns_5yr", x="category", data=dataset.sort_values("alpha", ascending=False), ax=axs[2])
axs[2].set_title("Returns (5 Years) by Category", fontsize=20)
```

## Returns (1 Year) by Category
# ![Returns (1 Year) by Category](returns_1yr_by_category.png)
# - The violin plot displays the distribution of 1-year returns by fund category.
# - Large-cap funds have the highest median returns, followed by mid-cap and small-cap funds.
# - However, the distribution of returns for small-cap funds is wider, indicating higher variability.

# Returns (3 Years) by Category
# ![Returns (3 Years) by Category](returns_3yr_by_category.png)
# - Similar to the 1-year returns, large-cap funds exhibit the highest median returns over a 3-year period.
# - However, the spread of returns for mid-cap funds is narrower compared to the 1-year returns.

# Returns (5 Years) by Category
# ![Returns (5 Years) by Category](returns_5yr_by_category.png)
# - Large-cap funds continue to have the highest median returns over a 5-year period.
# - The distribution of returns for small-cap funds remains wide, indicating higher risk.
# - Overall, investors may consider large-cap funds for consistent returns over different time periods.

# ## Conclusion
# - Large-cap funds consistently show higher median returns compared to mid-cap and small-cap funds.
# - However, investors should consider the risk associated with each category before making investment decisions.

```python
dataset_temp = dataset.groupby(['fund_age_yr'])['expense_ratio'].mean().reset_index()
plt.figure(figsize=(15, 6))
ax = sns.rugplot(x="fund_age_yr", y="expense_ratio", data=dataset_temp).set_title("Average Expense Ratio depending on Fund Age (Years)", fontsize=15)
ax = sns.stripplot(x="fund_age_yr", y="expense_ratio", data=dataset_temp)
```

## Average Expense Ratio by Fund Age
# ![Average Expense Ratio by Fund Age](expense_ratio_by_fund_age.png)
# - The rug plot displays the distribution of fund ages along the x-axis.
# - Each tick on the rug plot represents a fund with its respective age.
# - The strip plot overlays individual data points on top of the rug plot to show the distribution more clearly.
# - From the visualization, we can observe any patterns or trends in the relationship between fund age and average expense ratio.


# Conclusion
# - There appears to be some variability in the average expense ratio across different fund ages.
# - Further analysis may be needed to determine if there is a significant correlation between fund age and expense ratio.
 # - Understanding how expense ratios change over the lifespan of mutual funds can provide valuable insights for investors and fund managers.

```python
dataset_temp = dataset.groupby(['fund_age_yr'])['expense_ratio'].mean().reset_index()
plt.figure(figsize=(15, 6))
ax = sns.scatterplot(x="fund_age_yr", y="expense_ratio", data=dataset_temp).set_title("Average Expense Ratio depending on Fund Age (Years)", fontsize=15)
```
